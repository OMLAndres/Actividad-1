3. Genere un programa que una vez que haya leído los 15 usuarios debe de revisar que no haya nombres y boletas repetidas, en caso de ser así, se deben de borrar de la lista y los nombres y boletas resultantes deben de guardarse en un archivo usuarios_filtrados.txt

```
input_file='usuarios2.txt'
output_file='usuarios2.txt'

with open(input_file) as input:
    unique = set(line.rstrip('\n') for line in input.readlines())
with open(output_file, 'w') as output:
    output.write('\n'.join(unique))

import shutil, os
ruta = os.getcwd() + os.sep
origen = ruta + 'usuarios2.txt'
destino = ruta + 'usuarios_filtrados.txt'

if os.path.exists(origen):
    with open(origen, 'rb') as forigen:
        with open(destino, 'wb') as fdestino:
            shutil.copyfileobj(forigen, fdestino)
            print("Archivo copiado")
```
